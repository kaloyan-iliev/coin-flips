# Simulation to the coin toss rounds expectation

```python
import random
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
class coin():
    def __init__(self):
        self.count=0
    def flip(self):
        if random.randint(0,100)>49:
            print("HEADS")
        else:
            print("TAILS")
            self.count=1
```

```python
list1 = []
for i in range(10000):
    rnd=1
    coin1=coin()
    coin2=coin()
    coin3=coin()
    while not (coin1.count == 1 and coin2.count ==1 and coin3.count ==1):
        print(rnd)
        if coin1.count == 0:
            coin1.flip()
        if coin2.count == 0:
            coin2.flip()
        if coin3.count == 0:
            coin3.flip()
        print(rnd)
        rnd+=1
    list1.append(rnd)
```

np.mean(list1)
